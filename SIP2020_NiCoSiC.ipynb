Step3: **Shuffle** the entire dataset. Create **Y set** with dependent property (sound_pressure) and **X set** with independent variables (freq, angle of attack, chord length, free-stream vel, ss displacement thickness)

```
from sklearn.utils import shuffle

db = shuffle(db)
db
```

Step4: Create **training set** (X_train, Y_train) and **testing set** (X_test, Y_test) with train:test ratio as 80:20

```
features = ['Stirring (rpm)','I Density (A/dm2)', 'CoSO4 conc (g/L)', 'Particle in bath (g/L)'] #list
target = ['Hardness']
train_Set = db[:18]
Y_train = train_Set[target];
X_train = train_Set[features]

test_Set = db[18:]
Y_test = test_Set[target];
X_test = test_Set[features]
```

```
X_train
```



---

Step5: **Normalize** X_train and X_test using **min-max normalization i.e. X = (X-Xmin)/(Xmax-Xmin)**

Xmin and Xmax values are to be calcualted **for only training set** and **same values will be used for normalizing test set**


```
import numpy as np

Xmin = np.minimum(X_train["stirring"])
```

Step6: Perform linear regression (using scikit-learn library) and report error for both training and testing set

```
import sklearn
from sklearn import linear_model
```

```
Lin_reg = linear_model.LinearRegression();
Lin_reg.fit(np.array(X_train), np.array(Y_train));
print('Coeff:', Lin_reg.coef_);
print('Intercept:',Lin_reg.intercept_);

# CALCULATING mean absolute error: Training
mae_LinReg_train= np.mean(np.absolute(Lin_reg.predict(X_train)-Y_train));
per_err_train = np.mean((np.absolute(Lin_reg.predict(X_train)-Y_train))/Y_train)*100;
print('MAE_Linear Regression (Training set) = %.3f'%mae_LinReg_train)
print('Average Percent Error (Training set) = %.3f '%per_err_train+'%')


# TESTING (LEAST SQUARE REGRESSION)
pred_LinReg = Lin_reg.predict(X_test);
mae_LinReg_test= np.mean(np.absolute(pred_LinReg-Y_test));
per_err_test = np.mean((np.absolute(Lin_reg.predict(X_test)-Y_test))/Y_test)*100;
print('MAE_Linear Regression (Testing set) = %.3f'%mae_LinReg_test)
print('Average Percent Error (Testing set) = %.3f '%per_err_test+'%')
```

Step7: Create a parity plot for test set with actual values on X-axis and predicted values on Y-axis

```
plt.figure(figsize = [10,4])
plt.subplot(1,2,1)
plt.scatter(Y_train, Lin_reg.predict(X_train))
plt.xlim(0,700); plt.ylim(0,700); plt.plot([0,700],[0,700],color='r')
plt.xlabel('Actual'); plt.ylabel('Predicted(Lin Reg'); plt.title('Training: Parity Plot')

plt.subplot(1,2,2)
plt.scatter(Y_test, pred_LinReg)
plt.xlim(0,700); plt.ylim(0,700); plt.plot([0,700],[0,700],color='r')
plt.xlabel('Actual'); plt.ylabel('Predicted(Lin Reg'); plt.title('Testing: Parity Plot')
```
